## Ingest Folder of Videos

```python
import os
import glob
from datetime import datetime, timezone
from deeplake import Client
from dotenv import load_dotenv
from gemini_client import generate_video_caption, generate_text_embedding

load_dotenv()

# Initialize DeepLake
dl_client = Client()

# Use the exact table name your /search API endpoint is expecting
table_name = "ai_events" 

# Find all MP4 and MOV files in the motion_clips directory
video_files = glob.glob("motion_clips/*.mp4") + glob.glob("motion_clips/*.mov") + glob.glob("motion_clips/*.MOV")

print(f"Found {len(video_files)} videos to ingest.")

for video_path in video_files:
    print(f"\n--- Processing {video_path} ---")
    try:
        # 1. Grab the file's actual modification time to use as the true timestamp
        file_stat = os.stat(video_path)
        event_time = datetime.fromtimestamp(file_stat.st_mtime, tz=timezone.utc).isoformat()
        
        # 2. Upload to Gemini and get the caption
        caption = generate_video_caption(video_path)
        print(f"Caption: {caption}")
        
        # 3. Get the 768-D embedding vector
        embedding = generate_text_embedding(caption)
        
        # 4. Ingest into DeepLake (No schema parsing, pure data)
        dl_client.ingest(table_name, {
            "video_path": [video_path],
            "caption": [caption],
            "embedding": [embedding],
            "timestamp": [event_time]
        })
        
        print(f"✅ Successfully ingested {video_path}!")
        
    except Exception as e:
        print(f"❌ Failed to process {video_path}. Error: {e}")

print("\n🎉 Batch ingestion complete!")
```
